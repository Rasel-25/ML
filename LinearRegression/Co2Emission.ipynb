```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplot inline
```

# Data Gathering

```python
data=pd.read_csv('FuelConsumption.csv')
data.head()
```

```python
# Excluding MODELYEAR Column because all values are the same 
data.drop(columns=['MODELYEAR'],inplace=True)
```

# Statistical Analysis

```python
data.describe().T
```

# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
for col in ['MAKE','MODEL','VEHICLECLASS','TRANSMISSION','FUELTYPE']:
    data[col]=encoder.fit_transform(data[col])
```

# Exploratory Data Analysis (EDA)

```python
sns.pairplot(data)
```

```python
# seatter plot between ENGINSIZE and  C02EMISSIONS
plt.figure(figsize=(10,8))
plt.scatter(x=data['ENGINESIZE'],y=data['CO2EMISSIONS'],color='hotpink')
plt.xlabel('ENGINSIZE')
plt.ylabel('C02EMISSIONS')
plt.title('ENGINSIZE VS C02EMISSIONS')
plt.grid()
plt.show()
```

```python
# seatter plot between FUELCONSUMPTION_CITY and  C02EMISSIONS
plt.figure(figsize=(10,8))
plt.scatter(x=data['FUELCONSUMPTION_CITY'],y=data['CO2EMISSIONS'],color='green')
plt.xlabel('FUELCONSUMPTION_CITY')
plt.ylabel('C02EMISSIONS')
plt.title('FUELCONSUMPTION_CITY VS C02EMISSIONS')
plt.grid()
plt.show()
```

```python
# seatter plot between FUELCONSUMPTION_HWY and  C02EMISSIONS
plt.figure(figsize=(10,8))
plt.scatter(x=data['FUELCONSUMPTION_HWY'],y=data['CO2EMISSIONS'],color='m')
plt.xlabel('FUELCONSUMPTION_HWY')
plt.ylabel('C02EMISSIONS')
plt.title('FUELCONSUMPTION_HWY VS C02EMISSIONS')
plt.grid()
plt.show()
```

```python
sns.set(style="white")
sns.relplot(x=data['FUELCONSUMPTION_COMB'],y=data['CO2EMISSIONS'],data=data,height=7)
plt.grid()
plt.title('FUELCONSUMPTION_COMB VS C02EMISSIONS')
```
