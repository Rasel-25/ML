```python
import numpy as np 
```

```python
def gradient(x,y):
    m_curr=b_curr=0
    iterations=10000
    n=len(x)
    learning_rate=0.001
    for i in range(iterations):
        y_predicted=m_curr*x+b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
        md=-(2/n)*sum(x*(y-y_predicted)) # Derivative of m 
        md=-(2/n)*sum(y-y_predicted) # derivative of b 
        m_curr=m_curr - learning_rate*md # calculate step size of m
        b_curr=b_curr - learning_rate*md # calculate step size of b
        print("m {:.2f},         b {:.2f},          cost {},      iterations {}".format(m_curr,b_curr,cost,i))
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
gradient(x,y)
```
